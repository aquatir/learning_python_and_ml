```python
import numpy as np
import scipy.stats as stats
from matplotlib import pyplot

%matplotlib inline
```

```python
gamma_dist = stats.gamma(5, 2)
```

```python
x = np.linspace(0,15,100)
cdf = gamma_dist.cdf(x)

samples = gamma_dist.rvs(100)

from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(samples)

pyplot.plot(x, cdf, label="theoretical")
pyplot.step(ecdf.x, ecdf.y, label="ECDF")
pyplot.ylabel('$f(x)$')
pyplot.xlabel('$x$')
pyplot.legend(loc='lower right')
pyplot.show()
```

```python
pdf = gamma_dist.pdf(x)

pyplot.plot(x, pdf, label = "PDF")
pyplot.hist(samples, normed=True, label = "histogram")
pyplot.xlabel("$x$")
pyplot.ylabel("$F(x)$")
pyplot.legend(loc = "upper right")
pyplot.show()
```

```python
sample1 = gamma_dist.rvs(100)

for n in [250, 50,15,5,2]:
    samples_mean = 1000 * [0]
    for i in range (0,1000):
        sample = gamma_dist.rvs(n)
        samples_mean[i] = sample.mean()
    pyplot.hist(samples_mean, normed=True, label = "Values: " + str(n))
pyplot.legend(loc = "upper right")
pyplot.show()
```
